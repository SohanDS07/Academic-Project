```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Assuming your DataFrame is named df1

# Step 1: Calculate total crime for each state
df1['Total_Crime'] = df1.drop(['State/Ut', 'District', 'Year'], axis=1).sum(axis=1)

# Step 2: Get the top 5 states with the highest total crime count
top_5_states = df1.groupby('State/Ut')['Total_Crime'].sum().sort_values(ascending=False).head(5)

# Step 3: Visualize total crime for the top 5 states
plt.figure(figsize=(12, 6))
colors = sns.color_palette('Set2', len(top_5_states))

top_5_states_plot = top_5_states.plot(kind='bar', color=colors)
top_5_states_plot.set_ylabel('Total Crime Count')
top_5_states_plot.set_title('Top 5 States with Highest Total Crime Count')
plt.xticks(rotation=45)
plt.show()

# Step 4: Visualize the most common crime category in each of the top 5 states
crime_columns = df1.columns[3:-1]  # Exclude non-crime columns
for state in top_5_states.index:
    state_data = df1[df1['State/Ut'] == state]
    max_crime = state_data[crime_columns].idxmax(axis=1).values[0]
    max_crime_count = state_data.loc[:, max_crime].values[0]
    
    plt.figure(figsize=(8, 5))
    plt.bar(max_crime, max_crime_count, color='skyblue')
    plt.xlabel('Crime Category')
    plt.ylabel('Count')
    plt.title(f'Most Common Crime in {state}')
    plt.xticks(rotation=45)
    plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import zscore

# Assuming df1 is your DataFrame
# Replace this with your actual DataFrame
# df1 = pd.read_csv('path_to_your_file.csv')

# Sum the total crimes for each district
district_total_crimes = df1.groupby('District')['Total_Crime'].sum().reset_index()

# Calculate Z-scores for each district
district_total_crimes['Z_score'] = zscore(district_total_crimes['Total_Crime'])

# Define thresholds for anomalies (e.g., Z-score > 2 or Z-score < -2)
threshold = 2
anomalies = district_total_crimes[(district_total_crimes['Z_score'] > threshold) | (district_total_crimes['Z_score'] < -threshold)]

# Visualize the anomalies
plt.figure(figsize=(14, 8))
plt.bar(district_total_crimes['District'], district_total_crimes['Total_Crime'], color='lightgrey', label='Normal')
plt.bar(anomalies['District'], anomalies['Total_Crime'], color='red', label='Anomalies')
plt.xlabel('District')
plt.ylabel('Total Crime')
plt.title('Districts with Anomalously High or Low Crime Rates')
plt.xticks(rotation=90)
plt.legend()
plt.show()
```
