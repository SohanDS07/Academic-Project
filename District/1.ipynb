```python
import pandas as pd

# Read each file into separate DataFrames
df1 = pd.read_csv("01_District_wise_crimes_committed_IPC_2001_2012.csv")
df2 = pd.read_csv("01_District_wise_crimes_committed_IPC_2013.csv")
df3 = pd.read_csv("01_District_wise_crimes_committed_IPC_2014.csv")
```

```python
print("Columns of df1:")
print(df1.columns)

print("\nColumns of df2:")
print(df2.columns)

print("\nColumns of df3:")
print(df3.columns)
```

```python
# Remove rows where the district is labeled as "TOTAL" from df1 and df2
df1 = df1[df1['DISTRICT'] != 'TOTAL']
df2 = df2[df2['DISTRICT'] != 'TOTAL']
```

```python
# Concatenate df1, df2, and df3 into one DataFrame df
df = pd.concat([df1, df2], ignore_index=True)

# Display the first few rows of the combined DataFrame
print(df.head())
```
