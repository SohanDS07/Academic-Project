```python
import pandas as pd



# Define the regions
northern_states = ['Himachal Pradesh', 'Jammu & Kashmir', 'Punjab', 'Uttarakhand', 'Haryana']
northern_uts = ['Chandigarh', 'Delhi']

eastern_states = ['Arunachal Pradesh', 'Assam', 'Manipur', 'Meghalaya', 'Mizoram', 'Nagaland', 'Sikkim', 'Tripura', 'West Bengal', 'Bihar', 'Odisha', 'Jharkhand']
eastern_uts = ['A&N Islands']

southern_states = ['Andhra Pradesh', 'Karnataka', 'Kerala', 'Tamil Nadu', 'Telangana']
southern_uts = ['Lakshadweep', 'Puducherry']

western_states = ['Goa', 'Gujarat', 'Maharashtra', 'Rajasthan', 'Madhya Pradesh', 'Chhattisgarh']
western_uts = ['D&N Haveli', 'Daman & Diu']

# Function to determine the region
def get_region(state_ut):
    if state_ut in northern_states or state_ut in northern_uts:
        return 'North'
    elif state_ut in eastern_states or state_ut in eastern_uts:
        return 'East'
    elif state_ut in southern_states or state_ut in southern_uts:
        return 'South'
    elif state_ut in western_states or state_ut in western_uts:
        return 'West'
    else:
        return 'Central'

# Apply the function to create a new column
df_long['Region'] = df_long['State/Ut'].apply(get_region)

# Print the updated dataframe
print(df_long)

```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Assuming df is your existing dataframe

# Group by state and sum the crime counts
state_crime_counts = df.groupby('State/Ut')['Count'].sum().reset_index()

print("Crime Counts for Each State:")
print(state_crime_counts)
print()

# Filter data for Jammu and Kashmir, Madhya Pradesh, and Andhra Pradesh
jk_data = df[df['State/Ut'] == 'Jammu & Kashmir']
mp_data = df[df['State/Ut'] == 'Madhya Pradesh']
ap_data = df[df['State/Ut'] == 'Andhra Pradesh']

# Get top 5 districts with highest crime counts in Jammu and Kashmir
top_5_districts_jk = jk_data.groupby('District')['Count'].sum().nlargest(5).index

# Get top 5 districts with highest crime counts in Madhya Pradesh
top_5_districts_mp = mp_data.groupby('District')['Count'].sum().nlargest(5).index

# Get top 5 districts with highest crime counts in Andhra Pradesh
top_5_districts_ap = ap_data.groupby('District')['Count'].sum().nlargest(5).index

# Function to visualize top 5 crime types for a given district
def visualize_top_5_crimes(data, district_name):
    # Filter data for the current district
    district_data = data[data['District'] == district_name]
    
    # Group by type of crime and sum the crime counts
    district_crime_counts = district_data.groupby('Type_of_Crime')['Count'].sum().reset_index()
    
    # Sort the data by crime counts and select the top 5 crimes
    top_5_crimes = district_crime_counts.nlargest(5, 'Count')
    
    # Plotting
    plt.figure(figsize=(10, 6))
    sns.barplot(x='Count', y='Type_of_Crime', data=top_5_crimes, palette='viridis')
    plt.title(f'Top 5 Crimes in {district_name}', fontsize=15)
    plt.xlabel('Total Crime Count', fontsize=12)
    plt.ylabel('Type of Crime', fontsize=12)
    plt.xticks(rotation=45)
    plt.show()

# Visualize top 5 crime types for each state
for state, top_5_districts in [('Jammu & Kashmir', top_5_districts_jk),
                               ('Madhya Pradesh', top_5_districts_mp),
                               ('Andhra Pradesh', top_5_districts_ap)]:
    print(f"Top 5 Districts in {state}:")
    print(top_5_districts)
    print()
    
    for district in top_5_districts:
        visualize_top_5_crimes(df[df['State/Ut'] == state], district)
```

```python
def print_top_crimes_for_districts(data, state):
    # Filter data for the state
    state_data = data[data['State/Ut'] == state]
    
    # Get top 5 districts with highest crime counts in the state
    top_5_districts = state_data.groupby('District')['Count'].sum().nlargest(5).index
    
    # Loop through each district
    for district in top_5_districts:
        # Filter data for the current district
        district_data = state_data[state_data['District'] == district]
        
        # Group by type of crime and sum the crime counts
        district_crime_counts = district_data.groupby('Type_of_Crime')['Count'].sum().reset_index()
        
        print(f"\nTop 5 Crimes in {state}, {district}:")
        print(district_crime_counts.nlargest(5, 'Count'))

# Call the function for each state
states = ['Bihar','Jammu & Kashmir', 'Madhya Pradesh', 'Andhra Pradesh']

for state in states:
    print_top_crimes_for_districts(df, state)
```
