```python
import os
import pandas as pd

# Path to the directory containing CSV files
data_folder = r'C:\Users\sohan\Downloads\New project\Crime Data\Data'

# Get a list of all CSV files in the directory
csv_files = [file for file in os.listdir(data_folder) if file.endswith('.csv')]

# Create an empty list to store DataFrames
dfs = []

# Iterate through each CSV file, read it into a DataFrame, and append it to the list
for file in csv_files:
    file_path = os.path.join(data_folder, file)
    try:
        df = pd.read_csv(file_path, error_bad_lines=False)
        dfs.append(df)
    except pd.errors.ParserError as e:
        print(f"Error reading {file}: {e}")

# Concatenate all DataFrames into a single DataFrame
combined_df = pd.concat(dfs, ignore_index=True)

# Now combined_df contains all data from CSV files in the Data folder
print(combined_df)
```
